```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
v = [0,1,1,0,1,1,0,0,1,1,0]
```

```python
dim=100
vx = []
```

```python
for i in range(1,11):
    f = np.ones(dim)
    x = f*v[i]
    vx = np.concatenate((vx,x))
```

```python
plt.subplot(3,1,1)
plt.plot(vx)
```

```python
t = np.linspace(0,5,len(vx))
f1 = 2
f2 = 5
w1 = 2*np.pi*f1
w2 = 2*np.pi*f2
y1 = np.cos(w1*t*2)
y2 = np.cos(w2*t*2)
plt.subplot(2,1,1)
plt.plot(t,y1)
plt.subplot(2,1,2)
plt.plot(t,y2)
```

```python
mi = []
for i in range(0,len(vx)):
    if vx[i] == 0:
        zero = np.array([y1[i]])
        mi = np.concatenate((mi,zero))
    else:
        one = np.array([y2[i]])
        mi = np.concatenate((mi,one))
```

```python
plt.subplot(3,1,1)
plt.plot(t,mi)
```
